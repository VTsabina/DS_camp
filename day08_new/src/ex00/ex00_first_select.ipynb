## Subtable ##
5. Получите подтаблицу с помощью одного запроса, где:
    + Используются только "uid" и "datetime".
    + Используются только данные о пользователях (user_*), а не о администраторах.
    + Данные отсортированы по "uid" по возрастанию.
    + Колонка индекса — "datetime".
    + "datetime" конвертируется в DatetimeIndex.
    + Название DataFrame — pageviews.

6. Закройте соединение с базой данных.

```python
pageviews = pd.read_sql("SELECT uid, datetime FROM pageviews WHERE uid LIKE '%user%' ORDER BY uid;", conn, parse_dates=['datetime'], index_col='datetime')
print(pageviews.index)
```
