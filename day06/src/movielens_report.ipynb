##### Ratings.Movies

Оценим рейтинги, связанные с фильмами. Для начала определим количество оценок в год с помощью метода `dist_by_year()` класса `Ratings.Movies`. Мы видим, что лидером по количеству оценок стал **2000 год**. Также достаточно большое количество оценок было в 1996, 2007, 2017 году. Между этими точками наблюдаются периоды спада, однако можно говорить о достаточно равномерном распределении рейтингов и о стабильно высоких показателях с 2015 по 2018 год.

```python
%%time
mv.dist_by_year()
```

Теперь посмотрим на распределение самих оценок с помощью метода `dist_by_rating()`. Видим, что **минимальная оценка (0.5)** была поставлена 1370 раза, а **максимальная (5.0)** - 13211 раз. **Наиболее частые оценки** - 4.0 (26818), 3.0 (20047), 5.0 (13211). Обратим внимание, что, во-первых, пользователи склонны ставить оценки без дробной части, а во-вторых, они чаще ставят положительные оценки. Это, однако, не может быть свидетельством среднего качества фильмов, но может кое-что сказать о самих зрителях: вероятно, они чаще оценивают фильм, если он им понравился, если нет - оставляют без оценки. 

```python
%%time
mv.dist_by_rating()
```

Теперь выведем топ 10 самых оцениваемых фильмов, воспользовавшись методом `top_by_num_of_ratings()`. Видим, что **максимальное количество оценок** для фильма составляет **329**, причем всё это - фольмы, выпущенные до 2000 года, когда происходит резкий рост количества оценок. Это может иметь значение для статистики.

```python
%%time
mv.top_by_num_of_ratings(10)
```

Выведем топ 10 фильмов по средней оценке пользователей. Если наша теория верна, то все они будут оценены на 5.0 - таких оценок должно быть много. Как видно, данные, собранные методом `top_by_ratings()`, это подтверждают.

```python
%%time
mv.top_by_ratings(10)
```

Теперь рассмотрим один из фильмов с наибольшим количеством оценок - Forrest Gump (1994) (ID=356).  Узнаем, в какие годы эти оценки выставлляись и какова была средняя оценка для каждого года, с помощью метода `rating_trend()`. Таким образом в контексте рейтингов фильмов мы можем сделать вывод об общей стабильности (стационарности) количества оценок, при этом пользователи активнее оценивают фильмы, которые им понравились, и выставляют больше положительных оценок (>=3). Однако, чтобы составить полную картину, необходимо проанализировать рейтинги в контексте пользователей.
