## 2. Iterations ##
2. Итерации: во всех следующих подзадачах необходимо вычислить `fines / refund * year` для каждой строки. Создайте новый столбец с рассчитанными данными. Измерьте время выполнения, используя магическую команду `%%timeit` в ячейке.
+ Напишите функцию, которая перебирает DataFrame с помощью `for i in range(0, len(df))`, `iloc`, и `append()` для списка. Назначьте результат функции в новый столбец DataFrame.
+ Сделайте это с помощью `iterrows()`.
+ Сделайте с помощью `apply()` и лямбда-функции.
+ Используйте объекты Series из DataFrame.
+ Сделайте так же, как в предыдущем пункте, но с помощью `.values`.

```python
df['Calculation'] = pd.Series()
df['Fines'] = df['Fines'].str.replace(',', '').astype(float)
df['Refund'] = df['Refund'].astype(int)
df
```

```python
%%timeit
l = []
if len(l) > 0:
    l = []
for i in range(0, len(df)):
    l.append(df.iloc[i, 2] / df.iloc[i, 1] * df.iloc[i, 5])
df['Calculation'] = l
```

```python
%%timeit
for i, row in df.iterrows():
    row['Calculation'] = row['Fines'] / row['Refund'] * row['Year']
```

```python
%%timeit
df['Calculation'] = df.apply(lambda row: row['Fines'] / row['Refund'] * row['Year'], axis=1)
```

```python
%%timeit
df['Calculation'] = df['Fines'] / df['Refund'] * df['Year']
```

```python
%%timeit
df['Calculation'] = (df['Fines'].values / df['Refund'].values) * df['Year'].values
```

## 4. Downcasting ##
4. Даункастинг:
+ Выполните `df.info(memory_usage='deep')` и обратите внимание на тип данных и использование памяти.
+ Создайте копию вашего начального DataFrame в другой DataFrame `optimized_df`.
+ Сделайте даункастинг с `float64` до `float32` для всех столбцов.
+ Сделайте даункастинг с `int64` до минимального возможного числового типа.
+ Выполните `info(memory_usage='deep')` для вашего нового DataFrame. Обратите внимание на типы данных и использование памяти.

```python
df.info(memory_usage='deep')
```

```python
optimized_df = df.copy()

optimized_df['Fines'] = pd.to_numeric(optimized_df['Fines'], downcast='float')
optimized_df['Calculation'] = pd.to_numeric(optimized_df['Calculation'], downcast='float')

optimized_df['Refund'] = pd.to_numeric(optimized_df['Refund'], downcast='integer')
optimized_df['Year'] = pd.to_numeric(optimized_df['Refund'], downcast='integer')

optimized_df.info(memory_usage='deep')
```

## 5. Categories ##
5. Категории:
+ Измените тип колонок с `object` на `category`.
+ В этот раз проверьте использование памяти. Оно, вероятно, уменьшится в 2–3 раза по сравнению с исходным DataFrame.

```python
optimized_df['CarNumber'] = optimized_df['CarNumber'].astype('category')
optimized_df['Make'] = optimized_df['Make'].astype('category')
optimized_df['Model'] = optimized_df['Model'].astype('category')
optimized_df.info(memory_usage='deep')
```
