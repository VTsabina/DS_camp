## Task ##

## 1. Read ##
1. Read_csv:
   + Отфильтровать строки с индексами 2 и 3 с помощью аргумента skiprows. Мы знаем, что эти наблюдения были ложными.
   + Отфильтровать последние две строки с конца (footer) с помощью аргумента skipfooter. Мы знаем, что эти наблюдения тоже были ложными.
   + Присвоить следующим столбцам имена: datetime, user.
   + Использовать datetime как индексный столбец.

```python
df = pd.read_csv("../data/feed-views.log", sep='\t', skiprows=[2, 3], header=None, names=['datetime', 'user'], index_col='datetime', skipfooter=2, engine='python')
```

## 2. Rename ##

2. Переименовать datetime в date_time.

```python
df.index.rename('date_time', inplace=True)
df.head(2)
```

## 3. Write ##

3. To_csv:
   + Использовать ‘;’ в качестве разделителя.
   + Сохранить в файл с именем feed-views-semicolon.log.

```python
df.to_csv("../data/feed-views-semicolon.log", sep=';')
```
