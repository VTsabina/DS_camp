## 4. Car numbers aggregations ##
4. Выполните агрегации по номерам автомобилей:  
+ Отобразите номера автомобилей, сгруппированные по количеству штрафов в порядке убывания. Мы хотим найти тех, кто чаще всего нарушает закон.  
+ Выберите из исходного DataFrame все строки, соответствующие автомобилю с наибольшим количеством штрафов. Хотим рассмотреть подробнее.  
+ Отобразите номера автомобилей, сгруппированные по сумме штрафов в порядке убывания. Мы хотим найти тех, кто заплатил больше всего.  
+ Выберите из исходного DataFrame все строки, соответствующие автомобилю с наибольшей суммой штрафов. Хотим рассмотреть подробнее.  
+ Отобразите таблицу, которая отвечает на вопрос: «Есть ли номера автомобилей, связанные с разными моделями?»

```python
sorted_counts = df.groupby('CarNumber')['Fines'].count().sort_values(ascending=False)
sorted_counts
```

```python
df[df['CarNumber'] == sorted_counts.index[0]]
```

```python
sorted_ammounts = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
sorted_ammounts
```

```python
df[df['CarNumber'] == sorted_ammounts.index[0]]
```

```python
# Не учитывая null-значения
df.groupby('CarNumber')['Model'].nunique().sort_values(ascending=False)
```

```python
# Или более предметно, комбинируя изготовителя и модель, с учетом null-значений
unique_combinations = df[['CarNumber', 'Make', 'Model']].drop_duplicates()
result = unique_combinations.groupby('CarNumber').size()
result.sort_values(ascending=False)
# Для первого номера в одном случае указана модель, в другом - null. 
```
